```python
#https://dzone.com/articles/modern-python-web-scraping-using-multiple-librarie
from selenium import webdriver
from bs4 import BeautifulSoup
import urllib2, requests, os
```

```python
class Scraper(object):  
    
    def __init__(self, base_url, category, end_page, n_images, js = False):
        self.base_url = base_url
        self.category = category
        self.end_page = end_page
        self.js = js
        self.count = 0
        self.n_images = n_images
        

    def download_images(self,image_url_array):
        
        for image_url in image_url_array:
                
            try:
                #Modify image size request
                src = image_url['src']
                src = src[0:src.find('&w=')] + '&w=195&h=300&dpr=2&quality=60'

                #Download the image
                request = urllib2.Request(src)

                #Open the file
                img = urllib2.urlopen(request).read()

                #Save the file
                file_name = self.category + '/' + self.category + '-' + str(self.count) + '.jpg'
                with open (file_name, 'w') as f: f.write(img)

                #Update progress
                self.count += 1
                print('Number of images downloaded: ' + str(self.count))
            except:
                print(image_url)
                print('Could not download image')
            
        
    def scan_product_page(self,product_url):
        
        #Find product url
        full_product_url = 'https://shop.nordstrom.com' + product_url

        #Get the html of the product page
        product_page_soup = BeautifulSoup(requests.get(full_product_url).text, "html.parser")
        
        try: 
            #Get list with all the url for the product images
            image_url_array = product_page_soup.find('div', { 'class': 'content' }).findAll('img')

            #Download only the fisrt n images
            self.download_images(image_url_array[:self.n_images])
        except:
            print('Product link failed: ' + full_product_url)
        
    def run(self):

        #Initialize JS friendly scraper if needed
        if self.js:
            browser = webdriver.PhantomJS()

        #Initialize count
        self.count = 0

        #Create directory if not already exist
        if not os.path.exists(self.category):
            os.makedirs(self.category)

        #Loop through all Nordstrom pages
        for page in range(1,self.end_page+1): 

            #Product page
            url = self.base_url + str(page)
            print(url)
            #Create parsed soup objects
            if self.js:
                browser.get(url)
                list_page_soup = BeautifulSoup(browser.page_source, 'html5lib')
            else:
                list_page_soup = BeautifulSoup(requests.get(url).text, 'html.parser')
            
            list_of_product_url = list_page_soup.findAll('a', {'class':'npr-D2xob'})
            

            #For a given Nordstrom page, find the links to the product pages
            for product_url in list_of_product_url:
                
                #Process the product page
                self.scan_product_page(product_url['href'])


```

```python
#Define the inputs
base_url = 'https://shop.nordstrom.com/c/womens-casual-dresses?breadcrumb=Home%2FWomen%2FClothing%2FDresses%2FCasual&page='
category = 'womens-casual-dresses'
end_page = 34
n_images = 1
js = True

#Initialize the scraper
scraperObject = Scraper(base_url = base_url, category = category, end_page = end_page, n_images = n_images, js = js)

#Run the scraper
scraperObject.run()
```
